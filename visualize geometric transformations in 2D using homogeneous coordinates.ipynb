```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def translation(tx, ty):
  return(np.array(
      [[1,0,tx],
       [0,1,ty],
       [0,0,1],
      ]
      ))
```

```python
def scaling(sx, sy):
  return(np.array(
      [[sx,0,0],
       [0,sy,0],
       [0,0,1],
      ]
      ))
```

```python
initial_point = np.array([[2],[1],[1]])
```

```python
T = translation(2,3)
S = scaling(2,3)
```

```python
translated_point = T @ initial_point
scaled_point = S @ translated_point
```

```python
x1,y1 = initial_point[0,0], initial_point[1,0]
x2,y2 = translated_point[0,0], translated_point[1,0]
x3,y3 = scaled_point[0,0], scaled_point[1,0]
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x1, y1, color='red', label='Initial Point')
plt.scatter(x2, y2, color='blue', label='Translated Point')
plt.scatter(x3, y3, color='green', label='Scaled Point')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axvline(x=0, color='black', linestyle='-')
plt.axhline(y=0, color='black', linestyle='-')
plt.grid(True, alpha=0.5, linestyle='--')
plt.axis('equal')
plt.legend()
plt.show()
```
